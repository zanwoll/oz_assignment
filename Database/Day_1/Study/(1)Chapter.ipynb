데이터베이스란?
- (1) 데이터저장소란?
- 효율적인 데이터 관리
- 데이터를 통합하여 관리
- 데이터 누락 및 중복 방지

DBMS 특징 (ACID)
- 데이터 독립성 : 물리적, 논리적 독립성
- 데이터 무결성 : 데이터의 유효성 검사를 통해 데이터 무결성을 구현
- 데이터 보안성 : 계정관리, 권한 설정
- 데이터 일관성 : 데이터의 불일치성을 배제

RDBMS

스키마 : 데이터베이스의 구조를 정의하는 것 ( 테이블, 행, 열, 인덱스, 관계 등의 데이터베이스 구조를 명시

SQL : 데이터베이스에서 데이터를 관리하기 위해 사용되는 표준 프로그래밍 언어

인덱스(index) : 데이터베이스에서 데이터 검색 속도를 빠르게 하기 위해 사용되는 객체

데이터베이스 관리 시스템 : 데이터베이스를 생성하고 관리하는 소프트웨어

트랜잭션 : 데이터베이스의 상태를 변화시키는 하나의 작업 단위


- 단점
1. 성능 : 대용량 데이터 처리나 복잡한 쿼리에 대한 성능이 다른 모델에 비해 낮을 수 있음
2. 확장 어려움 : 수직적 확장이 한계가 있고, 수평적 확장은 복잡하고 비용이 많이 들 수 있다.
3. 유연성 부족 : 스키마 변경이 어려워서 요구사항 변경에 대응하기 어려울 수 있음
4. 복잡한 조인 : 많은 테이블 간의 조인은 성능에 부정적인 영향을 미칠 수 있습니다.
5. 고비용 : 서버 하드웨어 및 라이선스 비용이 상대적으로 높을 수 있음

1. 데이터베이스 : 구조화된 데이터의 집합
2. 테이블 : 데이터베이스 내에서 데이터가 저장되는 장소
3. 행/레코드 : 테이블 내의 개별 데이터 항목을 나타냄
4. 열/필드 : 테이블 내의 특정 카테고리 또는 데이터 유형을 나타냄

- 기본 키 : 테이블 내의 각 행을 고유하게 식별하는 열
- 외래 키 : 다른 테이블의 기본 키를 참조하는 열

[데이터베이스 스키마]

데이터베이스 스키마는 데이터베이스에서 데이터의 구조와 구성을 정의하는 청사진 또는 설계도 입니다

- 논리적 스키마 : 테이블 간의 관계, 각 테이블의 속성과 데이터 타입, 제약 조건 등
- 물리적 스키마 : 데이터의 실제 저장과 관련이 있는 부분으로, 디스크에 저장하는 방식 등
