NoSQL이란?

- 스키마 유연성 : NoSQL 데이터베이스는 스키마가 동적이거나 덜 제한적이어서 데이터 모델을 쉽게 조정하고 변경이 가능
- 분산처리 : NoSQL은 대부분 분산 데이터베이스로 설계되어 있어 대량의 데이터를 여러 서버에 분산하여 처리할 수 있다
- 높은 가용성과 확장성 : NoSQL은 높은 가용성과 확장성을 제공하여 시스템이 확장될 때 성능이 떨어지지 않도록 함
- 다양한 데이터 모델 : NoSQL은 다양한 데이터 모델을 지원합니다. 주요 모델로는 문서 지향, 키-값, 와이드 컬럼, 그래프 등이 있음

(1) 데이터 타입:
1. 정형 데이터 : 정해진 스키마에 따라 구조화된 데이터로, 관계형 데이터베이스나 스프레드시트, CSV와 같은 형식에서 사용됨
2. 반정형 데이터 : 일정한 구조를 가지고 있지만 전체적인 데이터 모델에 엄격하게 준수하지 않는 데이터로, XML, JSON, HTML, 로그와 같은 형식에서 사용됩니다.
3. 비정형 데이터 : 구조나 형태가 없으며, 연산이 어려운 데이터로, 영상, 이미지, 음성, 텍스트 등이 해당됨

(2) 종류:
1. 문서 지향 데이터베이스:
* NoSQL에서 가장 많이 사용되는 타입 중 하나
* 데이터를 JSON이나 BSON과 같은 문서 형식으로 저장하는 방식
* 여러 필드를 포함하고 있어 복잡한 구조의 데이터를 표현 가능

2. 키-값 데이터베이스:
* 키와 값을 쌍으로 데이터를 저장하는 방식
* 주로 캐시나 설정 데이터와 같이 단순한 데이터에 사용

3. 와이드 컬럼 데이터베이스:
* 열 기반의 데이터 모델을 사용하여 테이블을 열의 집합으로 표현
* Bigtable과 같은 모델로, 여러 개의 컬럼 패밀리를 가진 테이블로 데이터를 저장
* 각 로우는 다양한 수의 컬럼을 가질 수 있음

4. 그래프 데이터베이스:
* 개형 데이터(부모-자식관계)를 표현하기 위해 그래프 구조를 사용하는 데이터 모델
* 노드(Node)와 엣지(Edge)로 구성

5. 오브젝트 데이터베이스:
* 객체 지향 프로그래밍의 객체를 직접적으로 데이터베이스에 저장하고 관리
* 관계형 데이터베이스와는 다르게 객체 간의 관계, 상속, 다형성과 같은 객체 지향 프로그래밍의 개념을 그대로 지원

인덱스의 종류

1. B tree 인덱스 : 데이터베이스 시스템에서 가장 널리 사용되는 인덱싱 구조 중 하나
* 자료를 효율적으로 검색, 삽입, 삭제할 수 있도록 설계된 자기 균형 트리 구조입니다.
* 균형 트리 구조 : 모든 리프 노드가 같은 레벨에 있어 검색 시간이 일정함
* 다중 키 : 각 노드는 여러 개의 키를 저장할 수 있음
* 다중 자식 : 한 노드는 여러 자식 노드를 가질 수 있음
* 정렬된 키 : 키들은 항상 정렬된 상태로 유지됨

2. hash 인덱스 : 해시 테이블을 기반으로 하는 데이터베이스 인덱싱 방식. 키-값 쌍을 빠르게 검색 할 수 있도록 설계된 구조로, 특정 조건에서는 B-tree보다 더 효율적인 성능을 제공함
* O(1) 검색 시간 : 이상적인 경우 상수 시간에 데이터 접근이 가능함
* 해시 함수 사용 : 키 값을 해시 함수로 변환하여 저장 위치를 결정
* 버킷 구조 : 충돌을 처리하기 위해 버킷 개념을 사용함
* 메모리 효율성 : 특정 상황에서 B-tree보다 적은 메모리를 사용할 수 있음

3. 공간 인덱스 : 2차원 이상의 공간 데이터를 효율적으로 저장하고 검색하기 위한 특수한 인덱스 구조. 지리정보시스템(GIS), 위치 기반 서비스, 컴퓨터 그래픽스 등에서 널리 사용됨
* 다차원 데이터 처리 : 2차원, 3차원 등 다차원 공간 객체를 효율적으로 관리함
* 공간 쿼리 최적화 : 범위 검색, 포함 관계, 교차 여부 등 다양한 공간 연산을 지원
* 계층적 구조 : 대부분의 공간 인덱스는 공간을 계층적으로 분할하는 구조를 가짐
* 근접 데이터 접근성 : 공간적으로 가까운 객체들은 저장 공간에서도 가까이 위치하도록 설계됨

Cassandra : CQL = SQL과 유사한 구문을 가진 전용 쿼리 언어
1. 분산 아키텍처 : 단일 장애점 없이 여러 노드에 데이터를 분산 저장함
2. 선형적 확장성 : 노드를 추가하면 성능이 거의 선형적으로 향상됨
3. 고가용성 : 데이터 복제를 통해 노드 장애에도 서비스 중단 없이 운영됨
4. 튜닝 가능한 일관성 : 강한 일관성에서 최종 일관성까지 필요에 따라 조정 가능함
5. 컬럼 기반 데이터 모델 : 와이드 컬럼 스토어 (wide-column store) 방식을 사용함

데이터 모델:
* 키스페이스(Keyspace) : RDBMS의 데이터베이스와 유사한 최상위 컨테이너
* 테이블(Table) : 행과 열로 구성된 데이터 구조 (과거에는 '컬럼 패밀리'라고 불림)
* 행(Row) : 고유 키로 식별되는 데이터 집합
* 열(Column) : 이름, 값, 타임스탬프로 구성된 데이터 단위

Hbase :
1. 컬럼 지향 저장소 : 와이드 컬럼 스토어 형태의 데이터 모델 사용
2. 분산 아키텍처 : Hadoop 분산 파일 시스템(HDFS) 위에 구축된 분산 데이터 베이스
3. 고확장성 : 수평적 확장이 용이하며 대규모 데이터셋 처리 가능
4. 실시간 랜덤 접근 : 대용량 데이터에 대한 실시간 읽기/쓰기 지원
5. 강한 일관성 : Cassandra와 달리 강한 일관성 모델 제공

데이터 모델:
* 테이블(Table): 행과 열로 구성된 데이터 집합
* 행(Row): 고유한 행 키(Row Key)로 식별되며 사전순으로 정렬됨
* 컬럼 패밀리(Column Family): 관련 열들의 논리적 그룹
* 컬럼(Column): 컬럼 패밀리 내의 개별 데이터 필드
* 셀(Cell): 행 키, 컬럼 패밀리, 컬럼, 타임스탬프로 식별되는 값의 저장 단위
* 타임스탬프(Timestamp): 각 셀 값의 버전을 관리하는 시간 식별자

Google BigTable:
1. 대규모 확장성 : 수천 대의 서버에 걸쳐 페타바이트 규모의 데이터 처리 가능
2. 와이드 컬럼 스토어 : 희소 다차원 정렬 맵 구조
3. 짧은 지연 시간 : 밀리초 단위의 응답 시간 제공
4. 고성능 : 초당 수백만 건의 읽기/쓰기 작업 처리 가능
5. 자동 샤딩 : 데이터 분산과 부하 균형을 자동으로 관리

데이터 모델:
* 행 키(Row Key): 테이블의 각 행을 고유하게 식별하는 키, 사전식으로 정렬됨
* 컬럼 패밀리(Column Family): 관련 데이터를 함께 저장하는 논리적 그룹
* 컬럼 한정자(Column Qualifier): 컬럼 패밀리 내의 개별 데이터 필드
* 셀(Cell): 행 키, 컬럼 패밀리, 컬럼 한정자, 타임스탬프로 식별되는 개별 데이터 단위
* 타임스탬프(Timestamp): 각 셀의 여러 버전을 구분하는 시간 식별자

BigTable 아키텍처:
1. 물리적 구성 요소
* 마스터 서버(Master Server): 메타데이터 관리 및 태블릿 할당을 담당
* Google File System(GFS): 기반 스토리지 레이어 (Cloud Bigtable에서는 Colossus 사용)
* Chubby: 분산 락 서비스로 클러스터 조정을 담당

2. 논리적 구성 요소
* 태블릿(Tablet): 테이블을 행 키 범위로 분할한 단위, 분산 처리의 기본 단위
* SSTable(Sorted String Table): 불변의 정렬된 키-값 맵으로 디스크에 데이터 저장
* Memtable: 메모리 내 데이터 구조, 최근 쓰기 작업을 버퍼링

Vertica: NoSQL 데이터베이스가 아니라, 대규모 분석 워크로드를 위해 설계된 열지향 분석형 데이터베이스 관리 시스템(ADBMS) 이다.
1. 컬럼 스토어 아키텍처 : 데이터를 행 대신 열 단위로 저장하여 분석 쿼리 성능 최적화
2. 대규모 병렬 처리(MPP) : 여러 노드에 걸쳐 쿼리를 병렬로 처리
3. 고압축률 : 열 기반 저장으로 높은 데이터 압축 효율성 제공
4. SQL 표준 지원 : ANSI SQL 표준 쿼리 언어 지원
5. 확장성 : 페타바이트 규모의 데이터 처리 가능

아키텍쳐:
1. 물리적 아키텍처
* 노드 : 개별 서버 인스턴스, CPU/메모리/스토리지 자원 제공
* 클러스터 : 여러 노드로 구성된 Vertica 설치 단위
* 네트워크 메시 : 노드 간 통신을 위한 고속 네트워크 구성

2. 논리적 아키텍처
* 데이터베이스 : 스키마와 객체의 최상위 컨테이너
* 스키마 : 테이블 과 뷰 등을 그룹화하는 네임스페이스
* 프로젝션 : 물리적 저장 구조, 테이블 데이터의 실제 저장 방식 정의
* 세그먼트 : 프로젝션 내의 데이터 파티션

Apache Druid: 실시간 분석과 OLAP(Online Analytical Processing) 워크로드에 최적화되어 있음
