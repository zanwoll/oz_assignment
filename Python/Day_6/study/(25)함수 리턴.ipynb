```python
# 값 없이 리턴하는 경우

def oz_return():
    print("파이썬")
    print("재미있어")
    return
    print("거짓말하네")

oz_return()
```

```python
# 값과 함께 리턴하는 경우

def oz_return():
    return "파이썬 재미있엉"

result  = oz_return()
print(result)
```

```python
# 아무것도 리턴하지 않은 경우

def oz_return():
    return

result = oz_return()
print(result)
```

```python
# None : 아무것도 없다, bool()함수를 이용할 경우 False로 간주됨

bool(None)
```
