함수를 정의할 때 사용하는 매개변수의 개수와 호출 시 입력되는 매개변수는 동일해야 함.


```python
# 매개변수를 적게 입력한 경우
def oz(message, n):
    for i in range(n):
        print(message)

oz("oz코딩스쿨")
```

```python
# 매개변수를 많이 입력한 경우

def oz(message, n):
    for i in range(n):
        print(message)

oz("oz코딩스쿨", 5, 3)
```

# 함수 구조
def 함수이름(매개변수, *가변 매개변수):
    코드

```python
def oz(n, *values):
    for i in range(n):
        for value in values:
            print(value)
        print()

oz(3, "이렇게", "저렇게", "가변 매개변수는 요렇게", "신기하죠")
```

단, 가변 매개변수를 사용하기 위해서는 2가지 주의점이 있습니다.

* 가변 매개변수 뒤에는 일반 매개변수가 올 수 없습니다.
* 가변 매개변수는 하나만 사용할 수 있습니다.

```python
# 가변 매개변수 뒤에 일반 매개변수가 올 경우

def oz(*values, n):
    for i in range(n):
        for value in values:
            print(value)
        print()

oz(3, "이렇게", "저렇게", "가변 매개변수는 요렇게", "신기하죠")
```
