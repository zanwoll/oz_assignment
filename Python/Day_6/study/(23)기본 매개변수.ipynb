### print() 함수에 필요 매개변수



```python
def oz(value, n=2):
    for i in range(n):
        print(value)
oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")
```

```python
# 기본 매개변수의 값을 변경하는 방법
def oz(value, n=2):
    for i in range(n):
        print(value)
oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.", n=10)
```

```python
# 기본 매개변수 뒤에 일반 매개변수가 올 경우.
def oz(n=2, value):
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")
```

### 기본 매개변수의 역할은 무엇인가요? :

1. 함수를 정의할 때 매개변수에 기본값을 지정하여, 함수 호출 시 해당 매개변수의 값을 생략하거나 지정하지 않았을 경우 기본값이 사용되도록 하는 역할

1. 기본 매개변수의 뒤에 일반 매개변수가 올 수 없습니다.
