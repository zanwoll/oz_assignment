식별자는 무엇이고 사용하는 이유는 뭔가요? :
1. (식별자란?) 변수, 함수, 클래스, 모듈 등 객체를 식별하는데 사용되는 이름
2. (사용하는 이유) 변수, 함수 등에 의미있는 이름을 부여하여 코드의 가독성을 높임
