```python
def oz(start, end, multiple):
    for i in range(start, end, multiple):
        print(i, " ", end="")
    print()

oz(1, 10, 3)
oz(start=1, end=50, multiple=5)
oz(start=1, end=100, multiple=10)
```

```python
def oz(start=1, end=10, multiple=1):
    for i in range(start, end, multiple):
        print(i, " ", end="")
    print()

oz(1, 10, 3)
oz(end=50)
oz(end=100, multiple=10)
```
