__미션__
- x에 5를 할당하고, y에는 10을 할당
- x의 값을 제곱하고, 그 값을 z에 할당
- y의 값을 2로 나눈 후, 그 값을 a에 할당
- a와 z를 더한 값을 b에 할당
- b의 값이 x와 같은지 출력
-  b와 a를 문자열로 변환하여 합친 결과를 출력


<br>

```python
x = 5
y = 10
z = x ** 2
a = y // 2
b = a + z

print(b == x)

b_b = str(b)
a_a = str(a)

print(b_b + a_a)
```
