__[문제4]__

<br>

창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함


다음 점수가 입력되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요.

---

창의성 : 62
시장성 : 71
안정성 : 68
수익성 : 66
<br>

* 참고
(creativity ,
marketability ,
stability ,
profitability )

__Tips__
* 비교 연산자와 논리 연산자의 결과를 고려해보아요

```python
crea = 62
mark = 71
stab = 68
prof = 66

total = ((prof >= 60) and (crea >= 50) and (mark >= 50) and (prof + mark) / 2 > stab)

print(total)
```

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

### [동료 캐릭터와 능력치 정보]

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
int = 80

magic = int > 50

print(magic)
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
str = 80
hp = 100

war = (str >= 60) and (hp >= 80)

print(war)
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
speed = 80
str = 40

thief = (speed >= 70) or (str >= 50)

print(thief)
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
int = 70
hp = 70

heal = (int >= 60) and (hp >= 50)

print(heal)
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
magic = 120
war = 150
thief = 110
heal = 90

print(magic >= 100)
print(war >= 100)
print(thief >= 100)
print(heal >= 100)
```
